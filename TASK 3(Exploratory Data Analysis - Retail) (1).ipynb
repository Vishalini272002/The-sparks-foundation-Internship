```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

# VISHALINI

```python
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.isnull().sum()
```

Therefore, this shows that there is no NaN values present in the dataset.

```python
df.describe()
```

```python
cor=df.corr()
sns.heatmap(data=cor,annot=True,)
```

```python
df.drop(['Postal Code'],inplace=True,axis='columns')
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=df,order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

Highest number of buyers are form California 

```python
plt.title("Sale of products of diffent category")
df['Category'].value_counts().plot.pie(autopct="%.2f%%")
```

Highest number of orders are for office supplies

```python
plt.title("Sales on each region")
df['Region'].value_counts().plot.pie(autopct="%.2f%%")
```

So we can see a largest sales of product in western side, to that lowest sales of product in southern side

```python
sns.pairplot(df,hue="Region")
```

Realtion graph of various parameters againts each other, w.r.t. "Region".

```python
sns.countplot(x=df['Ship Mode'])
```

Most of the people are preferring the standard class of shipping mode.

```python
sns.pairplot(df,hue="Ship Mode")
```

```python
plt.figure(figsize=(12,6))
plt.barh('Sub-Category','Category', data=df, color='red')
plt.xlabel("Categories",fontsize=13)
plt.ylabel("Sub-categories", fontsize=13)
plt.title("Sub-Categories distributed w.r.t Category",fontsize=15)
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar()
plt.ylabel('Profit/Loss and sales')
plt.title("Sales verses profit in category of products", fontsize=15)
plt.show()
```

```python
plt.title('Category wise Sales in each Region')
sns.barplot(x='Category',y='Sales',data=df,hue='Region',order=df['Category'].value_counts().index,palette='rocket')
plt.xlabel("Category",fontsize=15)
plt.show()
```

```python
df.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(width=0.8,figsize=(15,9))
plt.ylabel('Profit/Loss and sales')
plt.title("Sales VS profit on sub-category of products", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.countplot( data=df, x='Sub-Category', hue='Region')
plt.title("Region Wise Sale of different products", fontsize=15)
```
